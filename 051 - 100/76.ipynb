```python
import datetime as dt
from itertools import combinations_with_replacement
import numpy as np
```

```python
INICIO = dt.datetime.now()
for N in range(2, 15):
    a = list(range(N))
    contador = len([1 for t in combinations_with_replacement(a, N) if t[0] == 0 and sum(t) == N]) + 1
    instante = str(dt.datetime.now() - INICIO)
    print(N, "\t", contador, '\t', instante)
```

```python
N = 10
p = np.array([[0]*N]*N)
for i in range(1, N):
    p[i][1] = 1
#p[1][1] = 1
p
```

```python
for n in range(2, N):
    for k in range(2, n+1):
        v1 = p[n-1][k-1]
        v2 = p[n-k][k]
        print(f'p[{n}][{k}] = p[{n-1}][{k-1}] ({v1}) + p[{n-k}][{k}] ({v2})')
        p[n][k] = v1 + v2 
    print("")
p
```
