```python
%%cython --compile-args=-O3 --force

from libc.math cimport sqrt
from array import array
from cpython cimport array as c_array
cimport cython


@cython.boundscheck(False)
@cython.wraparound(False)
cdef cy_es_primo_simple(unsigned int N):
    cdef unsigned int extremo, i
    es = True
    extremo = (N // 2)+1
    for i in range(3, extremo, 2):
        if N % i == 0:
            es = False
            break
    return es


#@cython.boundscheck(False)
#@cython.wraparound(False)
cdef cy_funcion(unsigned int N_):
    cdef int p, mitad, total, n_mayor
    cdef float multi, actual, mayor, phi
    multi = 0
    actual = 0
    mayor = 0
    n_mayor = 0
    primos = tuple([2] + [i for i in range(3, N_+1, 2) if cy_es_primo_simple(i)])
    total = len(primos)
    print("total:", total)
    print(primos[:10], "...", primos[-5:])
    for N in range(2, N_+1):
        if N % 1000 == 0:
            print(N, end='\t')
        multi = 1
        mitad = (N // 2) + 1
        for p in primos:
            if p > mitad:
                break
            if N % p == 0:
                multi *= (1 - (1/p))
        phi = multi * N
        actual = N / phi
        if mayor < actual:
            mayor = actual
            n_mayor = N
    print("mayor:", mayor)
    print("n_mayor:", n_mayor)
    return n_mayor


def funcion(N):
    return cy_funcion(N)
```

```python
%%time
resultado = funcion(10)
print("resultado = {}".format(resultado))
```
