```python
Let r be the remainder when (a−1)^n + (a+1)^n is divided by a^2.

For example, if a = 7 and n = 3, then r = 42: 63 + 83 = 728 ≡ 42 mod 49. 
    And as n varies, so too will r, but for a = 7 it turns out that r_max = 42.

For 3 ≤ a ≤ 1000, find ∑ r_max.
```

```python
%%time
suma = 0
for a in range(7, 100):
    r_max = 0
    for n in range(1, a*10):
        r = ( ((a-1)**n) + ((a+1)**n) ) % (a*a)
        if r_max < r:
            r_max = r
    suma += r_max
    print(a, r_max, end="\t\t")
    break
print("")
suma
```
